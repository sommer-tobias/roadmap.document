## Vereinfachungen und Annahmen

Ziel dieses Berichts ist eine Grobabschätzung der aktuellen Energieproduktion sowie das Festlegen eines Ausbauziels für jeden erneuerbaren Energieträger. Der Fokus liegt auf der Gesamtenergiesystembetrachtung und nicht auf einer detaillierten Ausarbeitung der einzelnen Technologien. Die Studie soll ausreichend genau sein, um strategische Fragestellungen, welche das Gesamtenergiesystem betreffen, beantworten zu können. Folgende Einschränkungen sind zu beachten:

1.	*Speicher*: Es werden Jahresenergiebilanzen gezeigt. Saisonalen Dynamiken und der Ausgleich dieser Dynamiken durch Speicher sind nicht explizit betrachtet. Durch die Abstützung auf EP2050+ werden Speicher jedoch implizit in dem Masse vorausgesetzt, wie sie im Szenario ZERO-Basis vorgesehen sind. Die Massnahme "Förderung Winterstromproduktion und saisonaler Speicher (KS-E2.3)" des Planungsberichts Klima und Energie [@PKuE2021] befasst sich explizit mit Speichern. 

2. *Importe*: Auch im Jahr 2050 werden rund 25 % der Energieträger noch importiert (im Vergleich zu den heutigen rund 75 %). Importiert werden in der Jahresbilanz insbesondere nachhaltige Biomasse, Wasserstoff und darauf basierende synthetische Brenn- und Treibstoffe.
